```python

# Apply skill tagging logic based on actual INR value (not in lakhs)
def assign_skill_tagging(ctc):
    if pd.isna(ctc):
        return None
    elif ctc < 700000:
        return "Base"
    elif 700000 <= ctc < 1200000:
        return "Niche"
    else:
        return "Super-niche"
```

```python
# Create a new Skill Tagging column
# Apply logic **only** where Skill Tagging is missing
df['Skill Tagging'] = df.apply(
    lambda row: assign_skill_tagging(row['Offered CTC']) if pd.isna(row['Skill Tagging']) else row['Skill Tagging'],
    axis=1
)

df.head(10)
```

```python
# Step 1: Compute observed hike % per Skill Tagging group
df['observed_hike_pct'] = (df['Offered CTC'] - df["Candidate's Current CTC (Pre TTL Last CTC)"]) / df["Candidate's Current CTC (Pre TTL Last CTC)"]

# Remove outliers and NaNs
df_filtered = df[
    (df['observed_hike_pct'].notna()) & 
    (df['observed_hike_pct'] > 0) & 
    (df['observed_hike_pct'] < 2)  # Ignore >200% hike as outliers
]

# Calculate median hike % per Skill Tagging group
median_hike_by_tag = df_filtered.groupby('Skill Tagging')['observed_hike_pct'].median().to_dict()

# Step 2: Define adjustment function for missing Current CTCs only
def infer_current_ctc_from_offered(row):
    current = row["Candidate's Current CTC (Pre TTL Last CTC)"]
    offered = row["Offered CTC"]
    skill_tag = row["Skill Tagging"]

    if pd.isna(current) and pd.notna(offered):
        # Use median hike% for this skill tag or fallback
        group_hike = median_hike_by_tag.get(skill_tag, 0.35)  # fallback to 35%
        try:
            return round(offered / (1 + group_hike))
        except ZeroDivisionError:
            return round(offered / 1.35)
    return current  # Leave unchanged if not missing

# Step 3: Apply only on missing values
df["Candidate's Current CTC (Pre TTL Last CTC)"] = df.apply(infer_current_ctc_from_offered, axis=1)

# Cleanup temporary column
df.drop(columns=['observed_hike_pct'], inplace=True)


```

```python
def parse_year_month(value):
    year = int(value)
    month_fraction = value - year
    month = int(round(month_fraction * 10))  # Because .4 means 4 months
    return year, month

def to_year_month(year, month):
    # Cap months between 0–11
    if month < 0:
        year -= 1
        month += 12
    elif month > 11:
        year += month // 12
        month = month % 12
    return round(year + (month / 10), 1)
```

```python
def adjust_relevant_experience(df):
    df = df.copy()
    total_exp_col = 'Total Experience (In Years)'
    relevant_exp_col = 'Relevant Experience (In Years)'

    # Shuffle for reproducible randomness
    shuffled = df.sample(frac=1, random_state=42)

    total_rows = len(df)
    n_25 = int(total_rows * 0.25)
    n_50 = int(total_rows * 0.5)

    indices_25 = shuffled.index[:n_25]
    indices_50 = shuffled.index[n_25:n_25 + n_50]
    indices_rest = shuffled.index[n_25 + n_50:]

    for idx in indices_25:
        y, m = parse_year_month(df.at[idx, total_exp_col])
        y -= 1
        df.at[idx, relevant_exp_col] = to_year_month(y, m)

    for idx in indices_50:
        df.at[idx, relevant_exp_col] = df.at[idx, total_exp_col]

    for idx in indices_rest:
        y, m = parse_year_month(df.at[idx, total_exp_col])
        y -= 1
        m -= 6  # 1.5 years = 1 year and 6 months
        df.at[idx, relevant_exp_col] = to_year_month(y, m)

    return df


```

```python
# Identify rows with missing Offer in Hand
missing_offer_mask = df['Offer in Hand'].isna()

# Split the missing rows into 50% zeros and 50% filled with computed values
missing_indices = df[missing_offer_mask].index
n_missing = len(missing_indices)
n_zero = n_missing // 2
n_fill = n_missing - n_zero  # remainder goes to filled

# Randomly shuffle indices to avoid any bias
np.random.seed(42)  # reproducibility
shuffled_indices = np.random.permutation(missing_indices)

# 1. Fill first half with 0
zero_indices = shuffled_indices[:n_zero]
df.loc[zero_indices, 'Offer in Hand'] = 0

# 2. Fill second half with computed values
fill_indices = shuffled_indices[n_zero:]

def generate_offer_in_hand(row):
    current = row["Candidate's Current CTC (Pre TTL Last CTC)"]
    offered = row["Offered CTC"]
    
    if pd.notna(current) and pd.notna(offered):
        low = current * 1.10  # 10% above current CTC
        high = offered * 0.95  # 5% below offered CTC
        
        if low < high:
            return round(np.random.uniform(low, high))
        else:
            return round((current + offered) / 2)  # fallback to mean if range is narrow
    return 0  # fallback if values are NaN
```

```python
# Apply logic to fill these indices
df.loc[fill_indices, 'Offer in Hand'] = df.loc[fill_indices].apply(generate_offer_in_hand, axis=1)

df.head(10)
```

```python
def determine_approval(offered_ctc):
    if pd.isna(offered_ctc):
        return None
    elif offered_ctc >= 2500000:
        return "Global Approval"
    elif offered_ctc >= 1500000:
        return "Local Approval"
    else:
        return "No Approval"

df['Approvals'] = df['Offered CTC'].apply(determine_approval)
```
